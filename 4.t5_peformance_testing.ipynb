```python
from rouge import Rouge
import os 
from datetime import datetime
import openpyxl
import pandas as pd
import re
```

```python
input_dir = "./Data/Generation/"
generated_column = "automatic_text"
target_column =  "original_text"    
results_path = "./Data/Results/{}.xlsx".format(datetime.now().strftime('%Y-%m-%d-%H-%M-%S'))
```

```python
model_name_lst = []
rouge_1_p_lst = []
rouge_1_r_lst = []
rouge_1_f_lst = []
rouge_2_p_lst = []
rouge_2_r_lst = []
rouge_2_f_lst = []
rouge_l_p_lst = []
rouge_l_r_lst = []
rouge_l_f_lst = []

rouge = Rouge()
for df in os.listdir(input_dir):
    if (df != ".ipynb_checkpoints"): 
       
        data = pd.read_csv(input_dir + df)
        data = data[data[target_column]!=""]
        
        if  type(data[generated_column].to_list()[0]) == str:
            automatic = data[generated_column].to_list()
        else:
            automatic = [sentence[0] for sentence in data[generated_column].to_list()]
            

        gold = data[target_column].to_list()
        scores = rouge.get_scores(automatic , gold, avg=True)


        model_name_lst.append(df)
        rouge_1_p_lst.append(scores["rouge-1"]["p"])
        rouge_1_r_lst.append(scores["rouge-1"]["r"])
        rouge_1_f_lst.append(scores["rouge-1"]["f"])
        rouge_2_p_lst.append(scores["rouge-2"]["p"])
        rouge_2_r_lst.append(scores["rouge-2"]["r"])
        rouge_2_f_lst.append(scores["rouge-2"]["f"])
        rouge_l_p_lst.append(scores["rouge-l"]["p"])
        rouge_l_r_lst.append(scores["rouge-l"]["r"])
        rouge_l_f_lst.append(scores["rouge-l"]["f"])

    
result_df = pd.DataFrame(
            {    
                'model': model_name_lst,
                'rouge-1_f_score': rouge_1_f_lst,
                'rouge-2_f_score': rouge_2_f_lst,
                'rouge-l_f_score':  rouge_l_f_lst,
                
                'rouge-1_precision': rouge_1_p_lst,
                'rouge-2_precision': rouge_2_p_lst,
                'rouge-l_precision': rouge_l_p_lst,
                
                'rouge-1_recall':  rouge_1_r_lst,
                'rouge-2_recall': rouge_2_r_lst,
                'rouge-l_recall':  rouge_l_r_lst,
                
            }
      )
result_df.to_excel(results_path)
```
